# HEART DISEASE DATA SET

```python
import pandas as pd

data = pd.read_csv("https://docs.google.com/spreadsheets/d/1ncf4KMRuQ_ilZRtixB1ExGyB0xJtFW-H/export?format=csv")
data
```

# Data Preparation

## Missing Values

```python
# Mencari Missing Value
missing_values = data.isnull().sum()

# Output missing values untuk setiap kolom
print(missing_values)
```

## Duplicate Value

```python
duplicated_values = data[data.duplicated()]

# Output nilai yang duplikat
duplicated_values
```

```python
#Menghapus Nilai Duplikat
data = data.drop_duplicates()
data
```

## Outlier Data

```python
# Menampilkan Outlier pada Data Numerik
data[['age','trestbps','chol','thalach','oldpeak','ca']].plot(kind='box', subplots=True, figsize=(20,5))
```

```python
#Mengahapus Data Outlier
P25 = data[['age','trestbps','chol','thalach','oldpeak','ca']].quantile(0.25)
P75 = data[['age','trestbps','chol','thalach','oldpeak','ca']].quantile(0.75)

# Menghitung rentang IQR
IQR = P75 - P25

# Menetapkan batas bawah dan batas atas
lower_bound = P25 - 1.5 * IQR
upper_bound = P75 + 1.5 * IQR

# Menghapus outlier
data_clean = data[((data[['age','trestbps','chol','thalach','oldpeak','ca']] >= lower_bound) & (data[['age','trestbps','chol','thalach','oldpeak','ca']] <= upper_bound)).all(axis='columns')]

# Menampilkan Boxplot yang Outliernya sudah dihapus
data_clean[['age','trestbps','chol','thalach','oldpeak','ca']].plot(kind='box', subplots=True, figsize=(20,5))
```

## Error Values

```python
# CEK ERROR VALUES DATA CATEGORICAL
data_clean[['sex','cp','restecg','exang','slope','thal','target']].astype(object).describe()
```

Kolom 'thal' memiliki 4 unique values, di mana sehsrusnya hanya ada 3, yaitu 1,2,3. Berikut adalah nilai 'thal' selain 1,2,3

```python
true_values=[1,2,3]
data_clean[~data_clean['thal'].isin(true_values)]
```

```python
#Menghapus data yang salah
true_values=[1,2,3]
data_clean = data_clean[data_clean['thal'].isin(true_values)]
data_clean
```

## Imbalance Data

```python
#Menampilkan distribusi kolom target
data_clean['target'].value_counts().plot(kind='bar')
```

Dapat terlihat bahwa data cukup balance

```python
data_clean.info()
```

# **EDA**

## Tipe Data

```python
# Pembagian feature berdasarkan tipe data (berdasarkan keterangan kolom pada kaggle)
numeric = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca']
categorical = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal', 'target']

# Memperbaiki tipe data categorical
data_fix = data_clean.copy()
data_fix[categorical] = data_clean[categorical].astype(object)

# Membuat tabel feature
features = pd.DataFrame({'Features': data_fix.columns.to_list()})

def num_or_cat(x):
    if x in categorical:
        return 'Categorical'
    return 'Numeric'

features['Tipe data'] = features['Features'].apply(num_or_cat)

# Menampilkan tabel
features
```

# **Feature Engineering**

## Menambah fitur kolom gender dengan mengubah (0-1) menjadi female, male

```python
# Ubah nilai 0 menjadi 'female' dan nilai 1 menjadi 'male' pada kolom 'Sex'
data_fix['Gender'] = data_fix['sex'].map({0: 'female', 1: 'male'})

# Tampilkan DataFrame hasil perubahan
data_fix.head()
```

## Hubungan antara gender dengan target

```python
# Ubah nilai 0 menjadi 'female' dan nilai 1 menjadi 'male' pada kolom 'Gender'
data_fe_sex = pd.get_dummies(data_fix, columns=['sex'])
data_fe_sex = data_fe_sex.rename(columns={'sex_0': 'female', 'sex_1': 'male'})

# Tampilkan DataFrame hasil peru
pd.DataFrame(data_fe_sex[['female', 'male']].astype(object).describe())
```

```python
data_fe_sex['female_disease'] = data_fe_sex['female'] & data_fe_sex['target']
pd.DataFrame(data_fe_sex['female_disease'].value_counts())
```


Berdasarkan hasil di atas, dapat terlihat bahwa jumlah perempuan yang memiliki penyakit jantung hanya 67 orang, sedangkan 194 lainnya antara pria atau tidak memiliki penyakit jantung


## Hubungan Gender (Male) dengan target

```python
data_fe_sex['male_disease'] = data_fe_sex['male'] & data_fe_sex['target']
pd.DataFrame(data_fe_sex['male_disease'].value_counts())
```

Berdasarkan hasil di atas, dapat terlihat bahwa jumlah laki-laki yang memiliki penyakit jantung 84 orang.

## Klasifikasi Cholestrol

```python
#Mendefinisikan fungsi untuk klasifikasi tipe kolestrol
def chol_type (x) :
  if x < 200:
    return 'Normal'
  if x < 240:
    return 'Borderline'
  return 'Bahaya'

#Menambah kolom baru, yaitu tipe kolestrol pada dataframe
data_fix['Tipe_Kolestrol'] = data_fix['chol'].apply(chol_type)
data_fix.head()
```

## Hubungan Kolestrol dengan 'Target'

```python
#Melihat jumlah per kategori
data_fix['Tipe_Kolestrol'].value_counts()
```

```python
data_abnrm_chol = data_fix.copy()
data_abnrm_chol['disease_abnrm_chol'] = ((data_abnrm_chol['Tipe_Kolestrol'] != 'Normal') & data_abnrm_chol['target'])
pd.DataFrame(data_abnrm_chol['disease_abnrm_chol'].value_counts())
```

Berdasarkan hasil di atas, terdapat 124 orang yang memiliki tipe kolestrol borderline dan bahaya yang mengidap penyakit jantung

## Mengelompokkan ca (number of major vessels)

```python
# Feature encoding
data_enc_ca = pd.get_dummies(data_fix, columns=['ca'])
ca_encoded = [f'ca_{i}' for i in data_fix['ca'].unique()]
data_enc_ca[ca_encoded] = data_enc_ca[ca_encoded].astype(int).astype(object)

pd.DataFrame(data_enc_ca[ca_encoded].describe())
```

Akan dilihat banyak pasien yang mengidap penyakit jantung berdasarkan kategori ca (number of major vessels) 'ca_2', yang artinya pasien memiliki Terdapat sumbatan atau penyempitan dalam dua pembuluh darah koroner besar

```python
data_disease_ca = data_fix.copy()
data_disease_ca['disease_ca'] = (data_enc_ca['ca_2'] & data['target']).astype(int).astype(object)
pd.DataFrame(data_disease_ca['disease_ca'].value_counts())
```

Sebanyak 7 orang yang mempunyai number of major vessels atau jumlah pembuluh darah yang mengalamani penyumbatan pada tingkat 2 (terdapat sumbatan atau penyempitan dalam dua pembuluh darah koroner besar) memiliki penyakit jantung.


## Mengelompokkan 'thal' / detak jantung

```python
# Feature encoding
data_enc_thal = pd.get_dummies(data_fix, columns=['thal'])
thal_encoded = [f'thal_{i}' for i in data_fix['thal'].unique()]
data_enc_thal[thal_encoded] = data_enc_thal[thal_encoded].astype(int).astype(object)

data_enc_thal[thal_encoded].describe(include=object)
```

Dapat terlihat bahwa mayoritas responden memiliki detak jantung normal

## Hubungan 'thal' normal dengan target

```python
data_disease_nrm_hr = data_fix.copy()
data_disease_nrm_hr['disease_nrm_hr'] = (data_enc_thal['thal_1'] & data['target']).astype(int).astype(object)
pd.DataFrame(data_disease_nrm_hr['disease_nrm_hr'].describe())
```

Dapat dibentuk feature ’disease_nrm_hr’ yaitu iya atau tidaknya seseorang dengan detak jantung normal memiliki penyakit jantung.
Namun dari hasil di atas, terlihat hampir semua responden tidak memiliki penyakit jantung dengan detak detak jantung normal


## Hubungan 'thal' tidak normal dengan target

```python
data_disease_abnrm = data_fix.copy()
data_disease_abnrm['disease_abnrm'] = ((data_enc_thal['thal_1'] == 0) & data['target']).astype(int).astype(object)
pd.DataFrame(data_disease_abnrm['disease_abnrm'].describe())
```

Untuk melihat apakah detak jantung normal berpengaruh kepada kesempatan seseorang terkena penyakit jantung, akan dibentuk feature ’disease_abnrm_hr’ yaitu iya atau tidaknya seseorang mengidap penyakit jantung dengan detak jantung tidak normal
Dari hasil di atas, terlihat bahwa lebih banyak responden mengidap penyakit jantung dengan detak tidak normal. Artinya kita dapat mengasumsikan bahwa seseorang dengan detak jantung tidak normal memiliki peluang lebih besar terkena penyakit jantung
